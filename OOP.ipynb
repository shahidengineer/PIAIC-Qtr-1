# Type of methohs Python

```python
class Student:
    
    #class variable
    school = "Usman Institue of Technology"
    
    def __init__(self,m1,m2,m3):
        self.m1=m1
        self.m2=m2
        self.m3= m3
   
    # instance methods    
    def avg(self):
        return (self.m1+self.m2+self.m3)/3
    
    @classmethod
    def getSchool(cls):
        return cls.school
   

    @staticmethod
    def info():
        print("This is student class dear")
    
        
s1 = Student(89,98,99) 
s2 = Student(98,98,99) 

#Calling an instance method
print(s1.avg())
# Calling a class method 
print(s1.getSchool())
print(Student.info())
```

# Inheritance in Python

```python
# Define a class A with two functions
class A:
    def feature1(self):
        print("Feature1 is working")
    def feature2(self):
        print("Feature2 is working")
a1=A()
a1.feature1()
a1.feature2()
```

```python
# Define another class with two functions
class B():
    def feature3(self):
        print("Feature3 is working")
    def feature4(self):
        print("Feature4 is working")

b1 = B()
b1.feature3()
b1.feature4()
```

```python
class B(A):# Just by passing class name of parent class
    def feature3(self):
        print("Feature3 is working")
    def feature4(self):
        print("Feature4 is working")
# class was accessing only class B functions
# Now class B inherits class A, it can access all the
# attributes and behaviours of class A
```

```python
b2 = B()
b2.feature3()
b2.feature4()
b2.feature1()
b2.feature2()
```

# Multi-level Inheritance

# Instance as a variable

```python
class Car():
   
    """A simple attempt to represent a car."""
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0
    
    def get_descriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()
    
    def read_odometer(self):
        print("This car has " + str(self.odometer_reading) + " miles on it.")
    
    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            return("You can't roll back an odometer!")
    
    def increment_odometer(self, miles):
        self.odometer_reading += miles
    
    def fill_gas_tank(self):
    
        return ("Gas Tank is full")
```

```python
car1 = Car("Honda","City",2020)
print(car1.get_descriptive_name())
```

```python
car1.read_odometer()
```

```python
car1.update_odometer(500)
car1.read_odometer()
```

```python
car1.increment_odometer(200)
car1.read_odometer()
```

```python
print(car1.fill_gas_tank())
```
